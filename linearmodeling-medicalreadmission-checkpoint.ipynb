```python

# One-hot encoding for categorical variables with more than two levels
lmdata = pd.get_dummies(lmdata,columns=['Gender', 'Complication_risk'], drop_first=True,dtype=int)

print(lmdata.head())
```

```python
# Add a constant to the model
X = lmdata[[ 'Age',  'VitD_levels',  'HighBlood','Stroke','Overweight','Arthritis','Diabetes','Anxiety','Asthma','TotalCharge','Gender_Male','Gender_Nonbinary','Complication_risk_Low','Complication_risk_Medium']]
X = sm.add_constant(X)
# Convert all columns to numeric, coerce errors to NaN
X = X.apply(pd.to_numeric, errors='coerce')

# Drop any columns that still have NaN values after conversion
X = X.dropna(axis=1)

# Check for constant columns and remove them
const_cols = [col for col in X.columns if X[col].nunique() == 1]
X = X.drop(columns=const_cols)
# Convert DataFrame to a NumPy array of floats
X_values = np.array(X, dtype=float)

# Compute and view VIF
vif = pd.DataFrame()
vif["variables"] = X.columns
vif["VIF"] = [variance_inflation_factor(X_values, i) for i in range(X.shape[1])]

# View results using print
print(vif)

```

```python
# Fit the regression model after ensuring there are no constant columns
Y = lmdata['TotalCharge']
#X = lmdata[[ 'Age',  'VitD_levels',  'HighBlood','Stroke','Overweight','Arthritis','Diabetes','Anxiety','Asthma','TotalCharge','Gender_Male','Gender_Nonbinary','Complication_risk_Low','Complication_risk_Medium']]
#Initialmodel = smf.OLS('Y ~ X',data=lmdata)
Initialmodel= smf.ols('TotalCharge ~ Age + VitD_levels + HighBlood + Stroke + Overweight + Arthritis + Diabetes + Anxiety + Asthma + Gender_Male + Gender_Nonbinary + Complication_risk_Low + Complication_risk_Medium', data=lmdata).fit()
print(Initialmodel.summary())
```

```python
# Backward stepwise elimination
def backward_stepwise_elimination(X, Y, significance_level=0.05):
    num_vars = len(X.columns)
    while num_vars > 0:
        model = sm.OLS(Y.astype(float), X.astype(float)).fit()
        coeffs = model.summary2().tables[1]
        max_p_value = coeffs['P>|t|'].max()
        
        if max_p_value > significance_level:
            high_pval_predictor = coeffs['P>|t|'].idxmax()
            print(f"Removing predictor: {high_pval_predictor} with p-value: {max_p_value}")
            X = X.drop(columns=[high_pval_predictor])
            num_vars -= 1
        else:
            break
    return X, model

# Perform backward stepwise elimination
X_refined, refined_model = backward_stepwise_elimination(X, Y)
print(refined_model.summary())
```

```python
finalmodel=refined_model
# Obtain AIC and BIC values for the final model
aic_finalmodel = finalmodel.aic
bic_finalmodel = finalmodel.bic
print(f"AIC for final model: {aic_finalmodel}")
print(f"BIC for final model: {bic_finalmodel}")

# Obtain AIC and BIC values for the initial model
aic_Initialmodel= Initialmodel.aic
bic_Initialmodel = Initialmodel.bic
print(f"AIC for initial model: {aic_Initialmodel}")
print(f"BIC for initial model: {bic_Initialmodel}")
```
